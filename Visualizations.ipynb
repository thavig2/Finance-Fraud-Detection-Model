```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# load the data 

fraud_data = pd.read_csv("C:/Users/thanu/Downloads/CSC 555/fraud_data/Fraudulent_E-Commerce_Transaction_Data_2.csv")
```

```python
#view summary statistics 
fraud_data.describe(include="all")
```

```python
fraud_data.head()
```

```python
# Histogram of Transaction Amount 

plt.hist(fraud_data["Transaction Amount"], bins=20, color="blue", alpha=0.7, edgecolor="black")
plt.xlabel("Transaction Amount ($)")  # X-axis label
plt.ylabel("Frequency")  # Y-axis label
plt.title("Histogram of Transaction Amounts")  # Plot title
plt.show()

'''From this plot, we can tell it is right skewed with most values for transaction amount within the range [0,2000]. The frequency shows us with 20000. 
'''
```

```python
# Bar Chart for Payment Method

plt.figure(figsize=(8, 5))
plt.bar(fraud_data["Payment Method"].value_counts().index, fraud_data["Payment Method"].value_counts(), 
        color="green", alpha=0.7, edgecolor="black")
plt.xlabel("Payment Method")  # X-axis label
plt.ylabel("Count")  # Y-axis label
plt.title("Payment Method Distribution")  # Plot title
plt.xticks(rotation=45)  # Rotate labels for better visibility
plt.show()

'''From this plot for Payment Method Distribution, we can see different payment methods are evenly distributed with about a count of 6000. '''
```

```python
# Bar Chart Non-Fraudulent vs Fraudulent Transactions

fraud_counts = fraud_data["Is Fraudulent"].value_counts()

plt.bar(["Non-Fraudulent", "Fraudulent"], fraud_counts, color=["blue", "red"], alpha=0.7, edgecolor="black")
plt.xlabel("Fraud Status")  # X-axis label
plt.ylabel("Count")  # Y-axis label
plt.title("Fraud vs Non-Fraud Transactions")  # Plot title
plt.show()

'''From this plot for Fraud vs Non-Fraud Transactions, we can tell there is a class imbalance when using the one of the csv files in our fraud detection data
(Fraudulent_E-Commerce_Transaction_Data_2.csv with 6MB). Most transactions are non-fraud transactions at 20000 and fraud transactions under 5000.'''
```

```python
#Histogram for Transaction Hour

plt.hist(fraud_data["Transaction Hour"], bins=24, color="purple", alpha=0.7, edgecolor="black")
plt.xlabel("Transaction Hour")  # X-axis label
plt.ylabel("Frequency")  # Y-axis label
plt.title("Transaction Distribution by Hour")  # Plot title
plt.show()

'''From this plot, Transaction by Distribution by Hour, it is evenly distributed as transactions are made every hour from range [0,20]. '''
```
